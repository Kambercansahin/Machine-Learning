```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns 
```

```python
df_athe=pd.read_csv("athlete_events.csv")
```

```python
df_athe
```

```python
plt.scatter(df_athe["Age"],df_athe["Height"],df_athe["Weight"])
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("X and Y Label")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Weight",y="Age",data=df_athe,hue="Sex")
plt.xlabel("Weight of athlete")
plt.ylabel("Age of Athlete")
plt.title("weight and Age")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Weight",y="Age",data=df_athe,hue="Medal")
plt.xlabel("Weight of Athelete")
plt.ylabel("Age of Athelete")
plt.title("Weight and Title")

plt.show()
```

```python
pandas_csv=pd.read_csv("athlete_events.csv")
```

```python
pandas_csv
```

```python
sns.scatterplot(data=pandas_csv,x="Height",y="Age",hue="Sex",style="Medal")
plt.xlabel("Height of Athlete")
plt.ylabel("Age of Athlete")
plt.title("Age and Height of Athlete")



plt.show()
```

```python
#line Plot
sns.set_style("whitegrid")
sns.lineplot(data=pandas_csv,x="Height",y="Age",hue="Medal")

plt.show()
```

```python
#Distiribition
sns.set_style("whitegrid")
sns.displot(data=pandas_csv,x="Height",hue="Medal")
```

```python
#Catplot
sns.catplot(data=pandas_csv,x="Medal",y="Height",hue="Sex",col="Season")
```

```python
#corelation
sns.heatmap(pandas_csv.corr(numeric_only=True),annot=True)
```

```python
data=np.array([5,6,7,8,9,10,40,50,60,70,40,200])
```
