```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#median,quartile(25,50,75),min max 1.50 * IQR,outlier
```

```python
data=np.array([10,15,20,25,30,35,40,45,50,55,60,100])
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=data)
plt.title("Box Plot")
plt.ylabel("Data Value")
plt.show()
```

```python
df=sns.load_dataset("titanic")
```

```python
df
```

```python
plt.subplot(1,2,1)
sns.boxplot(x="class",y="age",data=df)
plt.xlabel("Class")
plt.ylabel("Age")
plt.title("Age by Class")


plt.subplot(1,2,2)
sns.boxplot(x="class",y="fare",data=df)
plt.xlabel("Class")
plt.ylabel("Fare")
plt.title("Fare by Class")

plt.tight_layout()
plt.show()

```
