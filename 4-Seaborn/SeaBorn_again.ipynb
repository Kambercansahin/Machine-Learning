```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("athlete_events.csv")
df
```

```python
plt.scatter(x="Height",y="Weight",data=df)
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x=df["Height"],y=df["Weight"],hue="Sex",data=df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height and Weight")
```

```python
#LINE PLOT
```

```python
sns.set_style("whitegrid")
sns.lineplot(data=df,x="Height",y="Weight",hue="Sex")
```

```python
#Distiribition
```

```python
sns.set_style("darkgrid")
sns.displot(data=df,x="Height",hue="Sex")
```

```python
#Cat Plot
```

```python
sns.set_style("darkgrid")
sns.catplot(data=df,x="Medal",y="Height",hue="Season",col="Sex")
```

```python
#Corelation
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

```python
#Box Plot
```

```python
sns.boxplot(data=df,x="Medal",y="Age",hue="Sex")
```
