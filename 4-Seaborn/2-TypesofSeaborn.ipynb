```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_athe=pd.read_csv("athlete_events.csv")
```

```python
df_athe
```

```python
#LINE PLOT

sns.set_style("whitegrid")
sns.lineplot(data=df_athe,hue="Sex",x="Height",y="Weight")
plt.show()
```

```python
#Distirubition or Histogram ->DaÄŸÄ±lÄ±m
sns.set_style("whitegrid")
sns.displot(data=df_athe,x="Height",hue="Sex")
plt.show()
```

```python
#CatPlot ->categorized plot
sns.catplot(x="Medal",y="Height",hue="Sex",data=df_athe,col="Season")
plt.show()
#Seasona gÃ¶re iki taneye ayÄ±rdÄ±
```

```python
#Corelation
sns.heatmap(df_athe.corr(numeric_only=True),annot=True)
```
