```python
#Top 5 App in category
```

```python
df_app_category=df_clean.groupby(["Category","App"])["Installs"].sum().reset_index()
```

```python
df_app_category=df_app_category.sort_values("Installs",ascending=False)
```

```python
apps=["GAME","COMMUNICATION","TOOLS","PRODUCTIVITY","SOCIAL"]
plt.figure(figsize=(25,7))
for i,app in enumerate(apps):
    df_a=df_app_category[df_app_category["Category"]==app]
    df_a=df_a.head(5)

    
    plt.subplot(3,2,i+1)
    sns.barplot(x=df_a["Installs"],y=df_a["App"])
    plt.title(app,size=20)
plt.tight_layout()
plt.show()
```
