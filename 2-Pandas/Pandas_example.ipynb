```python
#APPLY
```

```python
df_apply=pd.read_csv('8-apply_function_data.csv')
```

```python
df_apply
```

```python
def Department(department):
    department=list(department)
    return department[0]
```

```python
df_apply["Short_Department"]=df_apply["Department"].apply(Department)
```

```python
df_apply
```

```python
def Performance(Experience):
    if Experience >10:
        return 1
    else:
        return 0
```

```python
df_apply["NewExperience"]=df_apply["Experience"].apply(Performance)
```

```python
df_apply
```

```python
df_apply["New_Performance_Score"]=df_apply["Performance_Score"]+df_apply["NewExperience"]
```

```python
df_apply
```

```python
#With Lambda
```

```python
df_apply["Formating"]=df_apply["Name"].apply(lambda x : x.replace("_"," "))
```

```python
df_apply
```
