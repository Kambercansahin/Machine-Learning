```python
numpy_array=np.array([10,2,30,40,50,60,70])
```

```python
numpy_array2=np.array([50,60,70,80,90,100,110])
```

```python
new_plt=plt.plot(numpy_array,numpy_array2,"g*-",label="Asasient")
new_plt2=plt.plot(numpy_array2,numpy_array,"r*-",label="Kasasient")
new_plt3=plt.plot(numpy_array,numpy_array*2,"b*-",label="**2")
new_plt=plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_array,numpy_array2,"g*-",label="1-2")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Plot")
plt.legend()
plt.subplot(1,2,2)
plt.plot(numpy_array2,numpy_array,"b*-",label="2-1")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Plot2")
plt.legend()
plt.show()
```

```python
#With Figure
```

```python
my_figure=plt.figure()
figure_axes=my_figure.add_axes([0.1,0.1,0.3,0.3])
figure_axes.plot(numpy_array,numpy_array2,"g*-",label="figure")

figure_axes.set_title("Figure")
figure_axes.set_xlabel("X axis")
figure_axes.set_ylabel("Y axis")
figure_legend=plt.legend()
plt.show()
```
