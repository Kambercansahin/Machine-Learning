```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data1=np.linspace(1,10,20)
```

```python
data2=np.linspace(10,20,20)
```

```python
#Scatter Diagram
```

```python
plt.scatter(data1,data2**2,color="black")
plt.xlabel("Data1")
plt.ylabel("Data2")
plt.title("Data1 and Data2")
plt.show()
```

```python
#Histogram
```

```python
new_arr=np.random.randint(0,100,50)
```

```python
plt.hist(new_arr,edgecolor="black")
plt.show()
```

```python
#box plot
```
