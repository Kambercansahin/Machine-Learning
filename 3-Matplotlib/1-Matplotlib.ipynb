```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list= [10,20,30,30,30,40,50,60,70,75]
```

```python
weight_list=[20,60,80,85,86,87,70,90,95,90]
```

```python
plt.plot(age_list,weight_list,"r") # y-->yellow,g,r,b
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
#with numpy
```

```python
np_age_list=np.array(age_list)
```

```python
np_weight_list=np.array(weight_list)
```

```python
plt.plot(np_age_list,np_weight_list,"y")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
numpy_arr1=np.linspace(0,10,20)#0 ile 10 arasnda 20 say ver demek
```

```python
numpy_arr2=numpy_arr1**3
```

```python
plt.plot(numpy_arr1,numpy_arr2,"b*-")
plt.show()
```

```python
#subplot
```

```python
plt.subplot(1,2,1)#1 row olsun 2 column olsun ve ben 1. grafii iziyorum anlamnda
plt.plot(np_age_list,np_weight_list,"y*-")
plt.subplot(1,2,2)#1 row ,2 column olsun ve ben 2. grafii iziyorum anlamnda
plt.plot(numpy_arr1,numpy_arr2,"r*-")
plt.show()
```

```python
#EKSEN EKLEME
```

```python
my_figure=plt.figure()
figure_axes=my_figure.add_axes([0.1,0.1,0.6,0.6])
figure_axes.plot(numpy_arr1,numpy_arr2,"g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("FIGURE TITLE")
plt.show()
```

```python
new_plt=plt.figure(dpi=100)
new_axes=new_plt.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(numpy_arr1,numpy_arr1**2,label="numpy_arr1**2")
new_axes.plot(numpy_arr1,numpy_arr1**3,label="numpy_arr1**3")
new_axes.legend()
plt.show()
```
