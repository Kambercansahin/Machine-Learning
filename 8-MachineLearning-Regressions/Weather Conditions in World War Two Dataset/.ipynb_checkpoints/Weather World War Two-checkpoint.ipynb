```python
#Add to mean(),median() or mode() for NaN value
print("Median Precip",df["Precip"].median())

print("Mean Precip",df["Precip"].mean())

print("Mode Precip",df["Precip"].mode()[0])
```

```python
df["PoorWeather"].unique()
```

```python
df["PRCP"].unique()
```

```python
df["PRCP"]=df["PRCP"].replace("T",np.nan)
```

```python
df["PRCP"]=df["PRCP"].astype(float)
```

```python
df["SNF"].unique()
```

```python
df["SNF"]=df["SNF"].replace("T",np.nan)
```

```python
df["SNF"]=df["SNF"].astype(float)
```

```python
df.info()
```

```python
#Add to mean(),median() or mode() for NaN value
print("Median PRCP,SNF",df["PRCP"].median(),df["SNF"].median())

print("Mean PRCP,SNF",df["PRCP"].mean(),df["SNF"].mean())

print("Mode PRCP,SNF",df["PRCP"].mode()[0],df["SNF"].mode()[0])
```

```python
add_to=["PRCP","SNF"]
for i in add_to:
    df[i]=df[i].fillna(df[i].median())
```

```python
df.info()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

```python
#MAX,MIN,MEA are Fahrenheit(F) , = MaxTemp,MeanTemp,MinTemp are Celsius(C)
df.drop(["MAX","MIN","MEA","Date"],axis=1,inplace=True)
```

```python
df.info()
```

```python
#Dependent and Independent Feature
X=df.drop(["MeanTemp","MinTemp"],axis=1)
y=df["MeanTemp"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression,ElasticNet,ElasticNetCV,Ridge,RidgeCV,Lasso,LassoCV
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=15)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(numeric_only=True),annot=True)
```

```python
#Standartization
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
y_pred=regression.predict(X_test)
```

```python
y_pred
```

```python
Score=r2_score(y_test,y_pred)
print(Score)
```

```python
#poly transformation
poly=PolynomialFeatures(degree=2,include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
regression_poly=LinearRegression()
regression_poly.fit(X_train_poly,y_train)
```

```python
y_pred_poly=regression_poly.predict(X_test_poly)
```

```python
Score_poly=r2_score(y_test,y_pred_poly)
print(Score_poly)
```

```python
#PIPLINE FOR POLY
from sklearn.pipeline import Pipeline
def poly_regression(degree):
    scaler=StandardScaler()
    regression_pipline=LinearRegression()
    poly_pipline=PolynomialFeatures(degree=degree,include_bias=True)
    pipeline=Pipeline([
        ("standard_scaler",scaler),
        ("poly_pipline",poly_pipline),
        ("poly_regression",regression_pipline)])
    pipeline.fit(X_train,y_train)
    score_pipline=pipeline.score(X_test,y_test)
    print("",score_pipline)
```

```python
for degree in [1,2,3,4]:
    poly_regression(degree)
```

```python
#LASSO and LASSOCV
lasso=Lasso()
lasso.fit(X_train,y_train)
y_pred_lasso=lasso.predict(X_test)
Score_lasso=r2_score(y_test,y_pred_lasso)
print("Score Lasso:",Score_lasso)
print(lasso.alpha)

lasscv=LassoCV(cv=5)
lasscv.fit(X_train,y_train)
y_pred_lassocv=lasscv.predict(X_test)
Score_lassocv=r2_score(y_test,y_pred_lasso)
print("Score LassoCv:",Score_lassocv)
print(lasscv.alpha_)
```

```python
#RIDGE and RIDGECV
ridge=Ridge()
ridge.fit(X_train,y_train)
y_pred_ridge=ridge.predict(X_test)
Score_ridge=r2_score(y_test,y_pred_ridge)
print("Score Ridge",Score_ridge)

ridgecv=RidgeCV(cv=5)
ridgecv.fit(X_train,y_train)
y_pred_ridgecv=ridgecv.predict(X_test)
Score_ridgecv=r2_score(y_test,y_pred_ridgecv)
print("Score RidgeCV",Score_ridgecv)
```

```python
#ElasticNet and ElasticNetCV
elastic=ElasticNet()
elastic.fit(X_train,y_train)
y_pred_elastic=elastic.predict(X_test)
Score_elastic=r2_score(y_test,y_pred_elastic)
print("Score Elastic",Score_elastic)

elasticcv=ElasticNetCV(cv=5)
elasticcv.fit(X_train,y_train)
y_pred_elasticcv=elasticcv.predict(X_test)
Score_elasticcv=r2_score(y_test,y_pred_elasticcv)
print("Score ElasticCv",Score_elasticcv)
```
