```python
import pandas as pd
import numpy as np
import matplotlib.pyplot
import seaborn as sns
```

```python
df=pd.read_csv("weatherHistory.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop("Formatted Date", axis=1,inplace=True)
```

```python
df.info()
```

```python
df["Summary"].unique()
```

```python
weather_order = [[
    "Clear",
    "Dry",
    "Partly Cloudy",
    "Breezy",
    "Breezy and Partly Cloudy",
    "Dry and Partly Cloudy",
    "Humid and Partly Cloudy",
    "Mostly Cloudy",
    "Breezy and Mostly Cloudy",
    "Dry and Mostly Cloudy",
    "Humid and Mostly Cloudy",
    "Overcast",
    "Humid and Overcast",
    "Breezy and Overcast",
    "Windy and Partly Cloudy",
    "Windy and Mostly Cloudy",
    "Windy and Overcast",
    "Breezy and Dry",
    "Windy and Dry",
    "Foggy",
    "Breezy and Foggy",
    "Windy and Foggy",
    "Drizzle",
    "Light Rain",
    "Rain",
    "Windy",
    "Dangerously Windy and Partly Cloudy"
]]
```

```python
#Label and ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder
```

```python
ord_encoder = OrdinalEncoder(categories=weather_order)
```

```python
df["Summary"]=ord_encoder.fit_transform(df[["Summary"]])
```

```python
df.head()
```

```python
df["Precip Type"].unique()
```

```python
df["Precip Type"].mode()[0]
```

```python
df["Precip Type"]=df["Precip Type"].fillna(df["Precip Type"].mode()[0])
```

```python
df["Precip Type"].unique()
```

```python
label_encode=LabelEncoder()
df["Precip Type"]=label_encode.fit_transform(df["Precip Type"])
```

```python
df.head()
```

```python
df["Precip Type"].unique()
```

```python
df.info()
```

```python
df.drop("Daily Summary",axis=1,inplace=True)
```

```python
df.info()
```

```python
#For linear regression
```

```python
X=df.drop("Summary",axis=1)
y=df["Summary"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
linear_regression=LinearRegression()
linear_regression.fit(X_train,y_train)
y_pred_linear=linear_regression.predict(X_test)
```

```python
from sklearn.metrics import r2_score
score_linear=r2_score(y_test,y_pred_linear)
print(score_linear)
```

```python
#FOR POLYNOMIAL REGRESSION
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3,include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)


poly_regression =LinearRegression()
poly_regression.fit(X_train_poly,y_train)
y_pred_poly=poly_regression.predict(X_test_poly)
score_poly=r2_score(y_test,y_pred_poly)
print(score_poly)
```

```python
#Lasso,LassoCV
from sklearn.linear_model import Lasso,LassoCV,Ridge,RidgeCV,ElasticNetCV,ElasticNet
lasso_regression=Lasso()
lasso_regression.fit(X_train,y_train)
y_pred_lasso=lasso_regression.predict(X_test)
score_lasso=r2_score(y_test,y_pred_lasso)
print("Lasso",score_lasso)

lassocv_regression=LassoCV(cv=5)
lassocv_regression.fit(X_train,y_train)
y_pred_lassocv=lassocv_regression.predict(X_test)
score_lassocv=r2_score(y_test,y_pred_lassocv)
print("Lassocv",score_lassocv)
```

```python
#Ridge,RidgeCV
ridge_regression=Ridge()
ridge_regression.fit(X_train,y_train)
y_pred_ridge=ridge_regression.predict(X_test)
score_ridge=r2_score(y_test,y_pred_ridge)
print("Ridge",score_ridge)


ridge_regressioncv=RidgeCV(cv=5)
ridge_regressioncv.fit(X_train,y_train)
y_pred_ridgecv=ridge_regressioncv.predict(X_test)
score_ridgecv=r2_score(y_test,y_pred_ridgecv)
print("Ridgecv",score_ridgecv)
```

```python
#ElasticNet vs ElasticElasticNetCV
elastic_regression=ElasticNet()
elastic_regression.fit(X_train,y_train)
y_pred_elastic=elastic_regression.predict(X_test)
score_elastic=r2_score(y_test,y_pred_elastic)
print("Elastic",score_elastic)

elastic_regressioncv=ElasticNetCV(cv=5)
elastic_regressioncv.fit(X_train,y_train)
y_pred_elasticcv=elastic_regressioncv.predict(X_test)
score_elasticcv=r2_score(y_test,y_pred_elasticcv)
print("Elastic",score_elasticcv)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import  accuracy_score,classification_report,confusion_matrix
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train,y_train)
y_pred_logistic=logistic_regression.predict(X_test)
accuracy_socre=accuracy_score(y_test,y_pred_logistic)
print(accuracy_socre)
print("---------------------------------")
classi_score=classification_report(y_test,y_pred_logistic)
print(classi_score)
print("---------------------------------")
confux_score=confusion_matrix(y_test,y_pred_logistic)
print(confux_score)
```

```python
#hyperparameter
logistic_regression=LogisticRegression()
```

```python
penalty=["l1","l2","elasticnet"]
c_value=[100,10,1,0.1,0.01]
solver = ["newton-cg","lbfgs","liblinear","newton-cholesky","sag","saga"]
```

```python
prams=dict(penalty=penalty,C=c_value,solver=solver)
```

```python
#grid search cv
from sklearn.model_selection import GridSearchCV,StratifiedKFold
```

```python
cv=StratifiedKFold()
grid_regression=GridSearchCV(estimator=logistic_regression,param_grid=prams,cv=cv,scoring="accuracy",n_jobs=-1)
```

```python
grid_regression.fit(X_train,y_train)
```

```python
grid_regression.best_score_
```

```python
y_pred_grid=grid_regression.predict(X_test)
accuracy_socre=accuracy_score(y_test,y_pred_grid)
print(accuracy_socre)
print("---------------------------------")
classi_score=classification_report(y_test,y_pred_grid)
print(classi_score)
print("---------------------------------")
confux_score=confusion_matrix(y_test,y_pred_grid)
print(confux_score)
```

```python
#Multiclass 
from sklearn.multiclass import OneVsOneClassifier,OneVsRestClassifier
```

```python
onevsone=OneVsOneClassifier(estimator=LogisticRegression())
onevsrest=OneVsRestClassifier(estimator=LogisticRegression())
```

```python
onevsone.fit(X_train,y_train)
y_pred_onevsone=onevsone.predict(X_test)
score=accuracy_score(y_test,y_pred_onevsone)
print("Score",score)
```

```python
onevsrest.fit(X_train,y_train)
y_pred_onevsrest=onevsrest.predict(X_test)
score=accuracy_score(y_test,y_pred_onevsrest)
print("Score",score)
```
