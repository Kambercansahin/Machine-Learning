```python
np.random.seed(42)
```

```python
df_titanic["deck"].value_counts()
```

```python
#upsampling
```

```python
from sklearn.utils import resample
```

```python
non_mising=df_titanic["deck"].dropna()
```

```python
fillvalues=resample(non_mising,replace=True,n_samples=df_titanic["deck"].isna().sum(),random_state=42).reset_index(drop=True)
```

```python
df_titanic["Deck"]=df_titanic["deck"].copy()
```

```python
df_titanic.loc[df_titanic["Deck"].isna(),"Deck"]=fillvalues.values
```
