```python
print(df_titanic["age"].mean())

df_titanic["age"].median()
```

```python
#mean imputation
df_titanic["Age_mean"]=df_titanic["age"].fillna(df_titanic["age"].mean())
```

```python
df_titanic[["age","Age_mean"]]
```

```python
#median imputation works if data is distributed normally but şf we have outliers then solution-->median imputation
df_titanic["Age_median"]=df_titanic["age"].fillna(df_titanic["age"].median())
```

```python
df_titanic
```

```python
df_titanic[["age","Age_mean","Age_median"]]
```

```python
df_titanic["Age_median"].isnull().sum()
```

```python
#mode imputation ->categorical values
```

```python
df_titanic[df_titanic["embarked"].isnull()]
```

```python
df_titanic["embarked"].unique()
```

```python
mode_value=df_titanic[df_titanic['embarked'].notna()]["embarked"].mode()[0]
```

```python
mode_value
```

```python
df_titanic["embarked_mode"]=df_titanic["embarked"].fillna(mode_value)
```

```python
df_titanic
```
